```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.exp(np.linspace(np.log(10**-(6)), np.log(10**(-8)), 10))
```

```python
x_extend = np.append(x, np.exp(np.linspace( np.log(10**(-8)), np.log(10**(-9)),4 ))[1:])
```

```python
y_acc_test = [100, 98.56, 100, 100, 94.0, 97.7499, 83.75, 67.0, 53.25, 47.25] + [50,50,50]
yerr_acc_test = [0, 2.87, 0, 0, 7.88, 3, 19.28, 14.845, 8.04, 5.72] +[0,0,0]
```

```python
y_acc_train = [100,98.56, 100, 100, 94.6875, 98.875, 86.125, 81.5, 79.875, 92.8125] + [50,50,50]
yerr_acc_train = [0,2.87, 0, 0, 7.36, 1.3778, 18.64, 17.975, 14.449, 2.9115] + [0,0,0]
```

```python
plt.rc('font', family='serif', size=16)
plt.rc('xtick', labelsize='x-small')
plt.rc('ytick', labelsize='x-small')

plt.figure(figsize = (10,8))
plt.errorbar(x_extend, y_acc_test, yerr=yerr_acc_test, marker='o', markersize=10, label='test')
plt.errorbar(x_extend, y_acc_train, yerr=yerr_acc_train, marker='o',  linestyle='dashed', markersize=10, label='train')
plt.xscale('log')
plt.xlabel('String Tension (G$\mu$)')
plt.ylabel('Accuracy')
plt.xlim(1.1*max(x_extend),0.9*min(x_extend))
plt.legend()
plt.minorticks_on()
plt.savefig('../figures/acc_vs_tension.jpg')
```
