```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x= np.array([80, 120, 200, 400, 600, 800])
x_1 = np.array([76,115,194,388,600, 800])
x_2 = np.array([78,121,200,398,590, 780])
x_3 = np.array([82,118,205,408,610, 820])
x_4 = np.array([80,124,210,418,620, 840])

y_1 = [68.75, 76.666, 81.5, 97.0, 99.0, 99.2] #10^-7
y_1err = [10, 19, 14, 5, 1, 1.2]

y_2 = [56.25, 53.3, 70.5, 95.8, 100, 99.6] #0.75 * 10^-7
y_2err = [7.9, 8.1, 11.1, 5.1, 1, 1]

y_3 = [51.25, 58.3, 67.00, 70.49, 80.66, 96.12] #0.5 * 10^-7
y_3err = [6.12, 8.7, 17.9, 19.01, 16.04, 6.85]

y_4 = [53.75, 53.3, 49.5, 62.3, 51.3, 52.5] #0.25 * 10^-7
y_4err = [10.89, 4.85, 7.81, 12.4, 1.72, 2.2]
```

```python
plt.rc('font', family='serif', size=16)
plt.rc('xtick', labelsize='x-small')
plt.rc('ytick', labelsize='x-small')

plt.figure(figsize = (14,8))
plt.errorbar(x_1, y_1, yerr=y_1err, marker='o', markersize=10, label=r'$G\mu = 10^{-7}$')
plt.errorbar(x_2, y_2, yerr=y_2err, marker='s', markersize=10, label=r'$G\mu = 7.5 \times 10^{-8}$')
plt.errorbar(x_3, y_3, yerr=y_3err, marker='*', markersize=18, label=r'$G\mu = 5.0 \times 10^{-8}$')
plt.errorbar(x_4, y_4, yerr=y_4err, marker='^', markersize=13, label=r'$G\mu = 2.5 \times 10^{-8}$')
plt.xscale('log')
plt.xlabel('Data Amount')
plt.ylabel('Accuracy')
xmin, xmax = plt.xlim()
plt.xlim(xmin, xmax)
plt.xticks(ticks=x, labels=x)
plt.hlines(50, xmin, xmax, color='black', label='Baseline', linestyle='dashed')
plt.legend(prop={'size': 14})
#plt.savefig('../figures/acc_vs_data_amount.eps', format='eps')
```
