```python
from scaling_solution_signal import generate_signal
from gaussian_fluctuations import generate_gaussian, get_Cls
```

```python
import matplotlib.pyplot as plt
import numpy as np
import copy
import matplotlib
```

We first generate the cosmic string signal using the "generate_signal()" function imported from scaling_solution_signal. The output is a 400x400 map. 

```python
map_signal = generate_signal(N=3, string_tension=10**(-5))
```

We then generate a gaussian background.

```python
map_noise = generate_gaussian() 
```

We then do stuff for the color scheme.

```python
my_cmap = 'RdBu'
```

### Scaling solution polarization signal of cosmic strings

```python
plt.rc('font', family='serif', size=16)
plt.rc('xtick', labelsize='x-small')
plt.rc('ytick', labelsize='x-small')

plt.figure(figsize=(10,10))
plt.imshow(map_signal, cmap=my_cmap)
plt.xticks(np.linspace(0,399,11),np.arange(0,11))
plt.yticks(np.linspace(0,399,11),np.arange(0,11))
plt.xlabel('Degrees')
plt.ylabel('Degrees')
cbar = plt.colorbar(fraction=0.046, pad=0.04)
cbar.set_label(r'Temperature ($\mu$K)', rotation=270, labelpad=25)
plt.title(r"Polarization Signal for a Scaling Solution of Cosmic Strings with $G\mu$=$10^{-7}$")
plt.tight_layout()
#plt.savefig('scaling_signal.jpg', format='jpg')
```

### Gaussian fluctuations of the polarization signal

```python
plt.rc('font', family='serif', size=16)
plt.rc('xtick', labelsize='x-small')
plt.rc('ytick', labelsize='x-small')

plt.figure(figsize=(10,10))
plt.imshow(map_noise, cmap=my_cmap)
plt.xticks(np.linspace(0,399,11),np.arange(0,11))
plt.yticks(np.linspace(0,399,11),np.arange(0,11))
plt.xlabel('Degrees')
plt.ylabel('Degrees')
cbar = plt.colorbar(fraction=0.046, pad=0.04)
cbar.set_label(r'Temperature ($\mu$K)', rotation=270, labelpad=15)
plt.title(r"Gaussian Noise")
plt.tight_layout()
#plt.savefig('gaussian_noise.jpg', format='jpg')
#plt.savefig('../../figures/gaussian_noise.eps', format='eps')
```
