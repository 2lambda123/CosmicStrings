```python
from scaling_solution_signal import generate_signal
from gaussian_fluctuations import generate_gaussian, get_Cls
```

```python
import matplotlib.pyplot as plt
import numpy as np
import copy
import matplotlib
```

We first generate the cosmic string signal using the "generate_signal()" function imported from scaling_solution_signal. The output is a 400x400 map. 

```python
map_signal = generate_signal(N=3, string_tension=3*10**(-7))
```

We then generate a gaussian background.

```python
map_noise = generate_gaussian() 
```

We then do stuff for the color scheme.

```python
my_cmap = copy.copy(matplotlib.cm.get_cmap('inferno')) # copy the default cmap
my_cmap.set_bad((0,0,0))
```

### Scaling solution polarization signal of cosmic strings

```python
plt.imshow(map_signal, cmap=my_cmap)
plt.colorbar()
plt.title(f"Polarization Signal for a \n Scaling Solution of Cosmic Strings \n with N = 3")
```

### Gaussian fluctuations of the polarization signal

```python
plt.figure(figsize=(14,14))
plt.imshow(map_noise, cmap=my_cmap)
plt.colorbar()
```

### Cosmic String signal + Fluctuations signal combined 

```python
mult_factor = 30
plt.figure(figsize=(14,14))
plt.imshow(mult_factor*map_signal+map_noise, cmap=my_cmap) #Note : Cosmic String signal rescaled
plt.colorbar()
```
